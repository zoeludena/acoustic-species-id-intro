```python
list_of_possible = []
for audio_id in to_keep:
    audio_id_data = data[data.get("AudioMothCode") == audio_id]
    for time in possible_times:
        indexes = audio_id_data[audio_id_data.get("Hours") == time].get("NewIndex").to_numpy()
        list_of_possible.append(random.choice(indexes))
```
